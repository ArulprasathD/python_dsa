```python
import array as ar

arr = ar.array("i",[1,2,3,4])

Even = 0
Odd = 0

for i in arr:
    if i % 2 ==0:
        Even += 1
    else:
        Odd += 1

print("EVEN :",Even)
print("ODD : ",Odd)
```

```python
import array as ar

arr = ar.array("i",[10,20,30,5,8])

first = second = -9999

for i in arr:
    if i > first:
        second = first
        first = i
    elif i > second and i != first:
        second = i

print("Second largest:", second)
```

```python
arr = ar.array('i', [1,2,2,3,3,3])

freq = {}

for i in arr:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1

for j in freq:
    print(j,"->",freq[j])
```

```python
arr = ar.array("i",[10,20,30])

def check(user):
    for i in arr:
        if user == i:
            print(f"Yes,{user} in list")
            break
        
    else:
            print(f"No ,{user} not in list")

user = int(input("enter number : "))

print([10,20,30])
check(user)
```
