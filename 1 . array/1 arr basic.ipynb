#  Access
#  insert
#  delete
#  search

```python
from array import *

arr = array('i',[1,2,3,4,5,6])

print(arr[3])
print(arr[4])
print(arr[0])
```

```python
from array import *

arr = array('i',[1,2,3,4,5,6])

arr.insert(9,8)

print(arr)
```

```python
from array import *

arr = array('i',[1,2,3,4,5,6])

arr.pop(4)

print(arr)
```

```python
from array import *

arr = array('i',[1,2,3,4,5,6])

a = 5

for i in arr:
    if i == a:
        print("true : ",i)
        break
```

```python
from array import *

arr = array('i',[1,2,3,4,5,6])

a = 5

for i in range(0,len(arr)):
    if arr[i] == a:
        print("index : ",i)
        break
```
